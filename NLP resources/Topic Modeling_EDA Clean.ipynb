### topic Modeling

 1. Latent Dirichlet Allocation:
* Probabilistic, generative model which uncovers topics latent to dataset 
* by assigning weights to words in corpus, 
* where each topic will assign different probability weights to each word.

```python
# Define helper function to print top words
def print_top_words(model, feature_names, n_top_words):
    for index, topic in enumerate(model.components_):
        message = "\nTopic #{}:".format(index)
        message += " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1 :-1]])
        print(message)
        print("="*70)
```

```python
# Putting all preprocessing steps together
# When you vectorize raw text with CountVectorizer, 
# dual stages of tokenizing and stopwords filtering are 
# automatically included as high-level component. 

# Here unlike NLTK tokenizer, Sklearn's tokenizer discards all single character terms 
# like ('a', 'w' etc) and also lower cases all terms by default. 

# Unfortunately, there is no built-in lemmatizer in vectorizer
# So we are left with couple of options. 
# 1. Implementing it separately everytime before feeding data for vectorizing 
# 1. Extending sklearn implementation to include this functionality. 

# Luckily for us, we have latter option 
# where we can extend CountVectorizer class by overwriting build_analyzer()
```

```python
# Extending CountVectorizer class with lemmatizer
```

```python
lemm = WordNetLemmatizer()
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))
    
# We have inherited and subclassed original Sklearn's CountVectorizer class 
# And we've overwritten build_analyzer() by implementing lemmatizer for each list in raw text matrix.
```

```python
# Storing the entire training text in a list
# Then calling our overwritten Count vectorizer
# c tf_vectorizer: LemmaCountVectorizer object
text = list(train.text.values)
# Calling our overwritten Count vectorizer
tf_vectorizer = LemmaCountVectorizer(max_df=0.95, 
                                     min_df=2,
                                     stop_words='english',
                                     decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
```

```python
print('tf',tf)
```

### 3b. Latent Dirichlet Allocation
Unsupervised learning algorithms. 
*There are different implements of LDA algorithm
* In this notebook, I will be using Sklearn's implementation. 
* Another very well-known LDA implementation is Radim Rehurek's gensim

Corpus - Document - Word : Topic Generation

*In LDA, modelling process revolves around 3 things: 
1. text corpus
2. its collection of documents,
3. D and words W in documents. 

Therefore algorithm attempts to uncover K topics from this corpus via following way (illustrated by diagram)


```python
# The LDA algorithm first models documents via mixture model of topics. 
# From these topics, words are then assigned weights based on probability distribution of these topics. 
# It is this probabilistic assignment over words that allow user of LDA to say 
# how likely particular word falls into topic. 
# Subsequently from collection of words assigned to particular topic, 
# are we thus able to gain insight as to what that topic may actually represent from lexical point of view.

# From standard LDA model, 
# there are few key parameters that we have to keep in mind 
# And you should consider programmatically tuning before we invoke model:
# n_components: number of topics which you specify to model
# $$$\alpha$$$ parameter: This is dirichlet parameter which can be linked to document topic prior
# $$$\beta$$$ parameter: This is dirichlet parameter which can be linked to topic word prior

# To invoke algorithm, we simply create LDA instance through Sklearn's LatentDirichletAllocation function. 
# various parameters would ideally have been obtained through some sort of validation scheme. 
# In this instance, optimal value of n_components (or topic number) was found 
# by conducting KMeans + Latent Semantic Analysis Scheme 
# whereby number of Kmeans clusters and number of LSA dimensions were 
# iterated through and best silhouette mean score.

lda = LatentDirichletAllocation(n_components=11, max_iter=5,
                                learning_method = 'online',
                                learning_offset = 50.,
                                random_state = 0)
```

```python
# Topics which are generated by LDA


lda.fit(tf)
```

```python
# We will utilize helper function print_top_words() to return top 10 words
# which are attributed to each of LDA generated topics. 
# To select number of topics, this is handled through parameter n_components in function.

n_top_words = 40
print("\nTopics in LDA model: ")
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
```
